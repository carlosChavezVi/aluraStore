# 2. Ventas por categoría

```python
import pandas as pd

def clasificar_departamentos(tienda, columna_categoria):
    """
    Lee un archivo CSV, identifica y lista los departamentos únicos
    de una columna específica.

    Args:
        nombre_archivo_csv (str): La ruta al archivo CSV (ej. 'tienda_1.csv').
        columna_categoria (str): El nombre de la columna que contiene
                                 las categorías de producto (ej. 'categoria de producto').

    Returns:
        list: Una lista de los departamentos únicos encontrados en la columna.
              Retorna None si el archivo no se encuentra o la columna no existe.
    """
    try:
        # 1. Cargar el archivo CSV en un DataFrame de pandas
        df = pd.read_csv(tienda)

        # 2. Verificar si la columna de categoría existe
        if columna_categoria not in df.columns:
            print(f"Error: La columna '{columna_categoria}' no se encontró en el archivo CSV.")
            print(f"Columnas disponibles: {df.columns.tolist()}")
            return None

        # 3. Obtener los valores únicos de la columna de categoría
        departamentos_unicos = df[columna_categoria].unique().tolist()

        # 4. Imprimir la información
        print(f"Archivo cargado: {tienda}")
        print(f"Columna analizada: '{columna_categoria}'")
        print(f"\nNúmero de departamentos diferentes: {len(departamentos_unicos)}")
        print("\nLista de departamentos diferentes:")
        for depto in sorted(departamentos_unicos): # sorted para una mejor legibilidad
            print(f"- {depto}")

        return departamentos_unicos

    except FileNotFoundError:
        print(f"Error: El archivo '{tienda}' no se encontró. Asegúrate de que esté en la ruta correcta.")
        return None
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
        return None

# --- Uso del código ---
# Asegúrate de que 'tienda_1.csv' esté en el mismo directorio
# y que la columna se llame exactamente 'categoria de producto' (o el nombre real).
nombre_csv = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
nombre_columna = 'Categoría del Producto' # ¡Importante! Asegúrate de que este sea el nombre exacto de la columna

lista_departamentos = clasificar_departamentos(nombre_csv, nombre_columna)

if lista_departamentos:
    print("\nProceso completado. La lista de departamentos únicos es:")
    print(lista_departamentos)
```

```python
import pandas as pd

def calcular_ventas_por_departamento(nombre_archivo_csv, columna_categoria, columna_precio):
    """
    Lee un archivo CSV, calcula el total de ventas por cada departamento
    basándose en las columnas de categoría y precio.

    Args:
        nombre_archivo_csv (str): La ruta al archivo CSV (ej. 'tienda_1.csv').
        columna_categoria (str): El nombre de la columna que contiene
                                 las categorías de producto (ej. 'categoria de producto').
        columna_precio (str): El nombre de la columna que contiene
                              el precio de cada producto (ej. 'precio').

    Returns:
        pd.Series or None: Una Serie de pandas con el total de ventas por departamento,
                           o None si hay un error (archivo no encontrado, columna no existe, etc.).
    """
    try:
        # 1. Cargar el archivo CSV en un DataFrame de pandas
        df = pd.read_csv(nombre_csv)

        # 2. Verificar si las columnas necesarias existen
        if columna_categoria not in df.columns:
            print(f"Error: La columna de categoría '{columna_categoria}' no se encontró en el archivo CSV.")
            print(f"Columnas disponibles: {df.columns.tolist()}")
            return None
        if columna_precio not in df.columns:
            print(f"Error: La columna de precio '{columna_precio}' no se encontró en el archivo CSV.")
            print(f"Columnas disponibles: {df.columns.tolist()}")
            return None

        # Asegurarse de que la columna de precio sea numérica.
        # Los errores se pueden deber a valores no numéricos (ej. texto, comas en lugar de puntos)
        df[columna_precio] = pd.to_numeric(df[columna_precio], errors='coerce')
        # Eliminar filas donde el precio no pudo ser convertido a numérico (se convirtió a NaN)
        df.dropna(subset=[columna_precio], inplace=True)

        # 3. Agrupar por la columna de categoría y sumar los precios
        ventas_por_departamento = df.groupby(columna_categoria)[columna_precio].sum()

        # 4. Imprimir los resultados
        print(f"Archivo cargado: {nombre_csv}")
        print(f"Columnas analizadas: '{columna_categoria}' y '{columna_precio}'")
        print("\n--- Total de ventas por departamento ---")
        print(ventas_por_departamento.sort_values(ascending=False).to_string()) # Ordenar de mayor a menor y mostrar completo

        return ventas_por_departamento

    except FileNotFoundError:
        print(f"Error: El archivo '{nombre_archivo_csv}' no se encontró. Asegúrate de que esté en la ruta correcta.")
        return None
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
        return None

# --- Uso del código ---
# Asegúrate de que 'tienda_1.csv' esté en el mismo directorio.
# ¡Importante! Revisa que los nombres de las columnas coincidan exactamente con tu CSV.
nombre_csv = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
nombre_columna_categoria = 'Categoría del Producto'
nombre_columna_precio = 'Precio
'

ventas_totales = calcular_ventas_por_departamento(nombre_csv, nombre_columna_categoria, nombre_columna_precio)

if ventas_totales is not None:
    print("\nCálculo de ventas completado.")
```
